```python
theaters_winter2019=soup.find_all('td', class_= 'a-text-right mojo-field-type-positive_integer')
```

```python
alltheaters_winter2019 = []
for thing in theaters_winter2019:
    alltheaters_winter2019.append(thing.text)
alltheaters_winter2019
```

```python
release_winter2019=soup.find_all('td', class_= 'a-text-left mojo-field-type-date a-nowrap')
```

```python
allrelease_winter2019 = []
for thing in release_winter2019:
    allrelease_winter2019.append(thing.text)
allrelease_winter2019
```

```python
distributor_winter2019=soup.find_all('td', class_= 'a-text-left mojo-field-type-studio')
```

```python
alldistributor_winter2019 = []
for thing in distributor_winter2019:
    alldistributor_winter2019.append(thing.text)
alldistributor_winter2019
```

```python
import pandas as pd
Winter2019 = pd.DataFrame({
        "title": all_titles_winter2019,
        "theaters": alltheaters_winter2019,
         "release": allrelease_winter2019,
        "distributor": alldistributor_winter2019,
        "gross": allgross_winter2019,
        "total_gross": total_gross_winter2019})

Winter2019['season'] = 'Winter 2019'
Winter2019
```
