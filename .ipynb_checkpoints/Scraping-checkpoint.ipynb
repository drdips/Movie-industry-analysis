```python
import requests
from bs4 import BeautifulSoup as BS
url="https://www.boxofficemojo.com/season/winter/2019/?ref_=bo_sl_table_2"
page = requests.get(url)
soup = BS(page.content, 'html.parser')
```

# Year 2019 - Movies by season

## Winter 2019: Runs Jan 4-Feb 28

```python
# Using find_all function to grab information regarding winter movie titles
titles_winter2019=soup.find_all('td', class_= 'a-text-left mojo-field-type-release mojo-cell-wide')
```

```python
# Getting all titles of winter movies
all_titles_winter2019 = []
for thing in titles_winter2019:
    all_titles_winter2019.append(thing.select('a')[0].string)
all_titles_winter2019
```

```python
# Grabbing all information regarding winter gross
gross_winter2019=soup.find_all('td', class_= 'a-text-right mojo-field-type-money mojo-estimatable')
```

```python
# Retrieving list of gross per winter movie
all_gross_winter2019 = []
for thing in gross_winter2019:
        all_gross_winter2019.append(thing.text)
allgross_winter2019 = all_gross_winter2019[::2]
allgross_winter2019
```

```python
# Retrieving list of total gross per winter movie
totalgross_winter2019 = []
for thing in gross_winter2019:
    totalgross_winter2019.append(thing.text)
total_gross_winter2019 = all_gross_winter2019[1::2]
total_gross_winter2019
```

```python
# Grabbing number of theaters movies were released at
theaters_winter2019=soup.find_all('td', class_= 'a-text-right mojo-field-type-positive_integer')
```

```python
# Retrieving list of number of theaters for winter movies
alltheaters_winter2019 = []
for thing in theaters_winter2019:
    alltheaters_winter2019.append(thing.text)
alltheaters_winter2019
```

```python
# Grabbing information regarding release date
release_winter2019=soup.find_all('td', class_= 'a-text-left mojo-field-type-date a-nowrap')
```

```python
# Retrieving list of release dates for winter movies
allrelease_winter2019 = []
for thing in release_winter2019:
    allrelease_winter2019.append(thing.text)
allrelease_winter2019
```

```python
# Grabbing information regarding distributors
distributor_winter2019=soup.find_all('td', class_= 'a-text-left mojo-field-type-studio')
```

```python
# Retrieving list of distributors for winter movies
alldistributor_winter2019 = []
for thing in distributor_winter2019:
    alldistributor_winter2019.append(thing.text)
alldistributor_winter2019
```

```python
# Create dataframe of all information regarding winter movies
import pandas as pd
Winter2019 = pd.DataFrame({
        "title": all_titles_winter2019,
        "theaters": alltheaters_winter2019,
         "release": allrelease_winter2019,
        "distributor": alldistributor_winter2019,
        "gross": allgross_winter2019,
        "total_gross": total_gross_winter2019})

Winter2019['season'] = 'Winter 2019'
Winter2019
```

## Holiday Season 2019: Runs Nov 1 - Jan 2, 2020

```python
# list all titles
titles_holiday2019=soup.find_all('td', class_= 'a-text-left mojo-field-type-release mojo-cell-wide')
all_titles_holiday2019 = []
for thing in titles_holiday2019:
    all_titles_holiday2019.append(thing.select('a')[0].string)
all_titles_holiday2019
```

```python
# list total gross per movie
total_gross_holiday2019 = []
for thing in gross_holiday2019:
        total_gross_holiday2019.append(thing.text)
totalgross_holiday2019 = total_gross_holiday2019[1::2]
totalgross_holiday2019
```

```python
# list number of theaters per movie
theaters_holiday2019=soup.find_all('td', class_= 'a-text-right mojo-field-type-positive_integer')

alltheaters_holiday2019 = []
for thing in theaters_holiday2019:
    alltheaters_holiday2019.append(thing.text)
alltheaters_holiday2019
```

```python
# combine all seasons to create year 2019 dataframe, change gross and total gross to integers, and split the season year into month and year
year2019 = [Winter2019, Spring2019, Summer2019, Fall2019, Holiday2019]
year_2019 = pd.concat(year2019)

year_2019['gross'] = year_2019['gross'].map(lambda x: int(x[1:].replace(',', '')))
year_2019['total_gross'] = year_2019['total_gross'].map(lambda x: int(x[1:].replace(',', '')))

year_2019['season_year'] = year_2019.season.map(lambda x: int(x[-4:]))
year_2019['seasonn'] = year_2019.season.map(lambda x: x[0:-4])
year_2019 = year_2019.drop('season', axis=1)
year_2019
```
